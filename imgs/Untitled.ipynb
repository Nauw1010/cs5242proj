```python
import matplotlib.pyplot as plt

n_traindata = ['$WBC_{1}$', '$WBC_{10}$', '$WBC_{50}$', '$WBC_{100}$']
acc = [0.923, 0.9687, 0.9855, 0.9896]
acc_mask = [0.945, 0.9699, 0.9867, 0.9902]

loss = [0.3068, 0.11, 0.0524, 0.0435]
loss_mask = [0.1938, 0.1048, 0.0512, 0.041]


_, axarr = plt.subplots(1,2, figsize=(9, 3))
axarr[0].title.set_text('Accuracy - w/ and w/o mask')
axarr[0].plot(n_traindata, acc, color='green', label='w/o mask')
axarr[0].plot(n_traindata, acc_mask, color='red', label='w/ mask')
axarr[0].legend()
axarr[0].set_ylabel('Accuracy')

axarr[1].title.set_text('Loss - w/ and w/o mask')
axarr[1].plot(n_traindata, loss, color='green', label='w/o mask')
axarr[1].plot(n_traindata, loss_mask, color='red', label='w/ mask')
axarr[1].legend()
axarr[1].set_ylabel('Loss')

plt.show()
```
